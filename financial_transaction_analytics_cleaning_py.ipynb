```python
import pandas as pd
import numpy as np

df = pd.read_csv('accounting_data.csv')
```

```python
df.info()
```

```python
# 1. Rename columns (Power BI-friendly, no spaces)
df.columns = [c.strip().replace(' ', '_').replace('(', '').replace(')', '')
              for c in df.columns]

df.info()
```

```python
# 2. Parse dates
df['Date']      = pd.to_datetime(df['Date'])
df['Month_Year']    = df['Date'].dt.to_period('M').astype(str)
df['Month']     = df['Date'].dt.strftime('%B')        # "January", "February" ...
df['Month_Num'] = df['Date'].dt.month                 # 1–12 for sort order in Power BI
df['Quarter']   = df['Date'].dt.quarter               # 1, 2, 3, 4
df['Year']      = df['Date'].dt.year                  # 2025, 2026 ...
df['DayOfWeek'] = df['Date'].dt.day_name()

df.head()
```

```python
# 3. Normalise boolean flags
df['Missing_Data_Indicator'] = df['Missing_Data_Indicator'].astype(int)
df['Transaction_Outcome']    = df['Transaction_Outcome'].astype(int)

df.info()
```

```python
# 4. Round floats to 4 dp for readability
float_cols = df.select_dtypes('float64').columns
df[float_cols] = df[float_cols].round(4)

df.head()
```

```python
# 5. Derived KPIs (add value for Power BI)
df['Net_Profit']    = df['Revenue'] - df['Expenditure']
df['OpEx_Ratio']    = (df['Operating_Expenses'] / df['Revenue']).round(4)
df['Success_Flag']  = df['Transaction_Outcome'].map({1:'Success', 0:'Failure'})

df.head()
```

```python
# 6. Validate — no negatives in Amount
assert (df['Transaction_Amount'] >= 0).all(), "Negative amounts found!"
```

```python
# 7. Export CSV
df.to_csv('accounting_clean.csv', index=False)
print(f"Exported {len(df)} rows, {len(df.columns)} columns")
```
